```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# b = x*/tanh(x*)
b = np.arange(0.1,5.1,0.5)
x = np.tanh(b)
```

```python
plt.figure(dpi=350)

plt.plot(b,x/b)
plt.show()
```

```python
# 0 = -x + b * tanh(x)
def f(b,x): return -x + b*np.tanh(x)
x = np.arange(-3,3,0.1)
```

```python
plt.figure(dpi=350)
b = np.arange(0,4,1)

for B in b:
    plt.plot(x,f(B,x),label = f"b={B}")

plt.legend()
plt.grid()
plt.show()
```
