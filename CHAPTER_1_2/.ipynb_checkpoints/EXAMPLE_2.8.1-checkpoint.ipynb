```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): return x * (1 - x)
```

```python
dt = 0.1
def k(x): return f(x) * dt
```

```python
def RungeKutta(x):
    k_1 = k(x)
    k_2 = k(x + k_1/2)
    k_3 = k(x + k_2/2)
    k_4 = k(x + k_3)
    return x + (k_1 + 2 * k_2 + 2 * k_3 + k_4) / 6
```

```python
plt.figure(dpi = 350)

LT,LX=10.0, 2.0
gridwidth=0.25
T, X= np.meshgrid(np.arange(0, LT, gridwidth), np.arange(0, LX,gridwidth))

U = T
V = f(X)

plt.quiver(T,X,U,V, np.arctan2(V, U),angles='xy',scale_units='xy', scale=10.0, width=0.002)

plt.xlim([0,LT])
plt.ylim([0,LX])

# 
# plt.grid()
plt.draw()
plt.show()
```

```python
def exact_f(x_0, t):
    return 1/(1 + x_0 * np.exp(-t))

plt.figure(dpi = 350)


LT,LX=10.0, 2.0
gridwidth=0.25
T, X= np.meshgrid(np.arange(0, LT, gridwidth), np.arange(0, LX,gridwidth))

U = T
V = f(X)

plt.quiver(T,X,U,V, np.arctan2(V, U),angles='xy',scale_units='xy', scale=10.0, width=0.002)


t = np.arange(0, 10, 0.1)
for x_0 in [ -1,  -0.4,0.1, 1, 10, 100]:
    plt.plot(t,exact_f(x_0,t))

plt.xlim([0,LT])
plt.ylim([0,LX]) 

# 
# plt.grid()
plt.draw()
plt.show()
```
